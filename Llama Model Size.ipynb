```python
# Install the required dependencies
!pip install whisper moviepy transformers scikit-learn psutil matplotlib

import os
import torch
import subprocess
import time
import psutil
from transformers import AutoModelForSeq2SeqLM, AutoTokenizer
from moviepy.editor import VideoFileClip
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
import matplotlib.pyplot as plt
import whisper  # OpenAI Whisper library

# Function to install missing dependencies
def install_if_missing(package, import_name=None):
    if import_name is None:
        import_name = package
    try:
        __import__(import_name)
    except ModuleNotFoundError:
        subprocess.check_call(["pip", "install", package])
        __import__(import_name)

# Install necessary dependencies
install_if_missing("moviepy")
install_if_missing("whisper")  # Corrected whisper package name
install_if_missing("matplotlib")
install_if_missing("transformers")
#install_if_missing("scikit-learn")  # Ensure scikit-learn is installed
install_if_missing("psutil")

# Load Whisper model for transcription
whisper_model = whisper.load_model("small")

# Replace LaMini with BART (open model) for summarization
MODEL_NAME = "facebook/bart-large-cnn"  # Open model for summarization
device = "cuda" if torch.cuda.is_available() else "cpu"
tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)
model = AutoModelForSeq2SeqLM.from_pretrained(MODEL_NAME).to(device)

# Extract audio from video and transcribe using Whisper
def extract_and_transcribe(video_path):
    audio_path = video_path.replace(".mp4", ".wav")
    video = VideoFileClip(video_path)
    video.audio.write_audiofile(audio_path, codec='pcm_s16le')

    result = whisper_model.transcribe(audio_path)
    transcript = result["text"]

    os.remove(audio_path)  # Cleanup audio file
    return transcript

# Summarize transcript using BART model
def summarize_bart(transcript):
    inputs = tokenizer("summarize: " + transcript, return_tensors="pt", max_length=512, truncation=True).to(device)
    summary_ids = model.generate(**inputs, max_length=250)
    return tokenizer.decode(summary_ids[0], skip_special_tokens=True)

# Compute model size in millions of parameters
def get_model_size():
    num_params = sum(p.numel() for p in model.parameters())
    return num_params / 1e6  # Convert to millions

# Compute output quality (cosine similarity between original and summary)
def compute_output_quality(original, summary):
    original_tokens = tokenizer(original, return_tensors="pt", padding=True, truncation=True, max_length=512).to(device)
    summary_tokens = tokenizer(summary, return_tensors="pt", padding=True, truncation=True, max_length=512).to(device)

    original_embeddings = model.base_model.get_input_embeddings()(original_tokens.input_ids).detach().cpu().numpy()
    summary_embeddings = model.base_model.get_input_embeddings()(summary_tokens.input_ids).detach().cpu().numpy()

    original_avg_embedding = original_embeddings.mean(axis=1)
    summary_avg_embedding = summary_embeddings.mean(axis=1)

    similarity = cosine_similarity(original_avg_embedding, summary_avg_embedding)
    return similarity.mean()  # Cosine similarity as output quality

# Compute computational efficiency (inference time and memory usage)
def compute_computational_efficiency(transcript):
    start_time = time.time()

    # Memory usage before inference
    process = psutil.Process(os.getpid())
    memory_before = process.memory_info().rss / 1024 / 1024  # In MB

    # Perform the inference (summary generation)
    summarize_bart(transcript)

    # Memory usage after inference
    memory_after = process.memory_info().rss / 1024 / 1024  # In MB

    end_time = time.time()

    # Inference time in seconds
    inference_time = end_time - start_time
    memory_usage = memory_after - memory_before

    # Normalize and combine inference time and memory usage as computational efficiency in percentage
    max_inference_time = 10  # Max inference time in seconds (normalize against this)
    max_memory_usage = 200  # Max memory usage in MB (normalize against this)

    normalized_inference_time = min(inference_time / max_inference_time, 1)
    normalized_memory_usage = min(memory_usage / max_memory_usage, 1)

    # Compute combined computational efficiency as the inverse of both (better if both are smaller)
    computational_efficiency = (1 - normalized_inference_time + 1 - normalized_memory_usage) / 2 * 100

    return computational_efficiency

# Process uploaded videos from Google Drive
video_folder = "/content/drive/MyDrive/video"
video_paths = [os.path.join(video_folder, f) for f in os.listdir(video_folder) if f.endswith(".mp4")]

# Store metrics for plotting
video_names = []
model_sizes = []
output_qualities = []
computational_efficiencies = []

for video_path in video_paths:
    print(f"Processing video: {video_path}")
    transcript = extract_and_transcribe(video_path)
    summary = summarize_bart(transcript)

    model_size = get_model_size()
    output_quality = compute_output_quality(transcript, summary)
    computational_efficiency = compute_computational_efficiency(transcript)

    video_names.append(os.path.basename(video_path))
    model_sizes.append(model_size)
    output_qualities.append(output_quality * 100)  # In percentage
    computational_efficiencies.append(computational_efficiency)

    print(f"Summary for {video_path}:")
    print(summary)
    print(f"Model Size: {model_size:.2f}M parameters")
    print(f"Output Quality (Cosine Similarity): {output_quality * 100:.2f}%")
    print(f"Computational Efficiency: {computational_efficiency:.2f}%")
    print("-" * 40)

# Plotting the results using matplotlib (line chart)
fig, ax = plt.subplots(figsize=(10, 6))

# Plot Model Size, Output Quality, and Computational Efficiency
ax.plot(video_names, model_sizes, label="Model Size (M Parameters)", marker='o', color='b')
ax.plot(video_names, output_qualities, label="Output Quality (%)", marker='o', color='g')
ax.plot(video_names, computational_efficiencies, label="Computational Efficiency (%)", marker='o', color='r')

# Adding labels and title
ax.set_xlabel('Videos')
ax.set_ylabel('Metrics (%)')
ax.set_title('Video Summarization Performance Metrics')
ax.legend()

# Rotate x-axis labels for better visibility
plt.xticks(rotation=45, ha="right")
plt.tight_layout()

# Show the plot
plt.show()
```
