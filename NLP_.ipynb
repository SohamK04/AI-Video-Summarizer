```python
import matplotlib.pyplot as plt
import numpy as np

# Data points
parameters = [250, 730, 1300]  # in millions
performance = [52, 58, 63]  # Average performance

# Baseline models
llama_performance = 58
alpaca_performance = 63

# Plot LaMini-Flan-T5 results
plt.figure(figsize=(6,5))
plt.plot(parameters, performance, 'o-', color='red', label='LaMini-Flan-T5')

# Horizontal reference lines
plt.axhline(y=llama_performance, color='black', linestyle='dotted', label='LLaMA-7B')
plt.axhline(y=alpaca_performance, color='green', linestyle='dashed', label='Alpaca-7B')

# Annotations
plt.text(parameters[0], performance[0]-1.5, 'LaMini-Flan-T5', fontsize=10)
plt.text(parameters[-1], performance[-1], 'Alpaca-7B', fontsize=10)
plt.text(parameters[1], llama_performance, 'LLaMA-7B', fontsize=10)

# Labels and legend
plt.xlabel("# of parameters (in millions)")
plt.ylabel("Average Performance")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.ylim(50, 70)
plt.xlim(0, 1500)

# Show the plot
plt.show()
```
